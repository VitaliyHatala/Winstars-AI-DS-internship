```python
from main import MnistClassifier
from tensorflow.keras.datasets import mnist
from sklearn.metrics import classification_report
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

classifier_cnn = MnistClassifier("cnn")
classifier_cnn.train(X_train, y_train)
```

```python
classifier_nn = MnistClassifier("nn")
classifier_nn.train(X_train, y_train)
```

```python
classifier_rf = MnistClassifier("rf")
classifier_rf.train(X_train, y_train)
```

```python
predictions_cnn = classifier_cnn.predict(X_test)
predictions_nn = classifier_nn.predict(X_test)
predictions_rf = classifier_rf.predict(X_test)
```

```python
print(classification_report(predictions_cnn,y_test))
```

```python
print(classification_report(predictions_nn,y_test))
```

```python
print(classification_report(predictions_rf,y_test))
```
